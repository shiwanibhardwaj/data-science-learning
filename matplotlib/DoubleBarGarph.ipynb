```python
import matplotlib.pyplot as plt
x=['python','java','c++','javascript']
y=[60,80,70,50]
c=[20,40,60,80]
plt.xlabel("languages")
plt.ylabel("percentile")
plt.title("graph")
plt.scatter(x,y,c=c,cmap="coolwarm",s=200,edgecolors='black',label="popularity",alpha=0.4)
plt.legend()
plt.colorbar(label='Color Intensity')
plt.xticks(rotation=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=['python','java','c++','javascript']
y=[60,80,70,50]
z=[40,50,60,20]
c=['r','y','g','m']
c1=["y","m","r","b"]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languages")
plt.ylabel("percentile")
plt.title("popularity of languages")

plt.bar(p,y,width,color=c,label="2020")
plt.bar(p1,z,width,color=c1,label="2025")

plt.legend()
plt.xticks(p+width/2,x,rotation=20)
plt.show()
```
