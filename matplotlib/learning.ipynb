```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[2,4,6,8]
c=['r','b','y','g']
plt.xlabel('numbers',fontsize=10)
plt.ylabel('square',fontsize=10)
plt.title("barchart",fontsize=20)
plt.bar(x,y,width=0.4,color=c,label='popular',align='edge',edgecolor='m',linewidth=2,linestyle=":",alpha=0.4)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['python','c','c++','java']
y=[2,4,6,8]
z=[4,8,10,14]
c=['r','b','y','g']
x_index=np.arange(len(x))
width=0.4
plt.xlabel('numbers',fontsize=10)
plt.ylabel('square',fontsize=10)
plt.title("barchart",fontsize=20)
plt.bar(x_index-width/2,y,width=width,color=c,label='popular',align='edge',edgecolor='m',linewidth=2,linestyle=":",alpha=0.4)
plt.bar(x_index+width/2,z,width=0.4,color='m',label='popular1' )
plt.xticks(x_index,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[2,4,6,8]
color=[10,30,50,60]
sizes=[10,2,30,51]
plt.title("scatter",fontsize=10)
plt.scatter(x,y,s=sizes,marker="*",c=color,cmap='viridis')
t=plt.colorbar()
t.set_label("colorbar")
plt.show()
```
